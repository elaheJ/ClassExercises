```python

plt.figure(figsize=(10,10))
plt.xlabel('days')

ax1 = myData.Cideal.plot(color='blue', grid=True, label='work completed - ideal')
ax2 =myData.Cactual.plot(color='red', grid=True, secondary_y=True, label='work completed - actual')
ax3 = myData.Lideal.plot(color='green', grid=True, label='work left - ideal')
ax4 =myData.Lactual.plot(color='orange', grid=True, secondary_y=True, label='work left - actual')

h1, l1 = ax1.get_legend_handles_labels()
h2, l2 = ax2.get_legend_handles_labels()
h3, l3 = ax3.get_legend_handles_labels()
h4, l4 = ax4.get_legend_handles_labels()

plt.legend(h1+h2, l1+l2, loc=2)
plt.show()
```

```python
fig, (ax1,ax2)= plt.subplots(1,2, figsize=(40,10))

ax1.set_xlabel('days')
ax2.set_xlabel('days')

#they both show up on the same chart
# work completed under idea scenario
ax1.plot(myData.Cideal, color='pink',label='work completed - ideal') 
ax3=ax1.twinx()
# work completed under actual scenario
ax3.plot(myData.Cactual,color='green', label='work completed - actual')

ax2.plot(myData.Lideal, color='pink',label='work to be done - ideal') 
ax4=ax2.twinx()
ax4.plot(myData.Lactual, color='green', label='work to be done - actual')
ax1.set_title('burn up')
ax2.set_title('burn down')

h1, l1 = ax1.get_legend_handles_labels()
h3, l3 = ax3.get_legend_handles_labels()
ax3.legend(h1+h3,l1+l3,loc=2, prop={'size':20})
h2, l2 = ax2.get_legend_handles_labels()
h4, l4 = ax4.get_legend_handles_labels()
ax4.legend(h1+h3,l1+l3,loc=2, prop={'size':20})


plt.show()
```
