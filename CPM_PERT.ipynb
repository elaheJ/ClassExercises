```python
#CPM PERT using the randomly generated data from Monday

import pandas as pd

simulationDF=pd.read_csv("/content/drive/MyDrive/496/simulationDataset.csv")
simulationDF.head(2)

```

```python
paths=['ADEJ','BCDEJ','BCFGJ','BCFHJ','BI']
pathsDF=pd.DataFrame(columns=paths)

ADEJ=[]
BCDEJ=[]
BCFGJ=[]
BCFHJ=[]
BI=[]
for  index, row in simulationDF.iterrows():
  ADEJ.append(row['A']+row['D']+row['E']+row['J'])
  BCDEJ.append(row['B']+row['C']+row['D']+row['E']+row['J'])
  BCFGJ.append(row['B']+row['C']+row['F']+row['G']+row['J'])
  BCFHJ.append(row['B']+row['C']+row['F']+row['H']+row['J'])
  BI.append(row['B']+row['I'])

pathsDF['ADEJ']=ADEJ
pathsDF['BCDEJ']=BCDEJ
pathsDF['BCFGJ']=BCFGJ
pathsDF['BCFHJ']=BCFHJ
pathsDF['BI']=BI

pathsDF.head(2)



```

```python
import numpy as np
criticalPath=[]
criticalPathLength=[]
floatCritcialPathLength = np.array(criticalPathLength, dtype = np.float32)

pathCounts=[0,0,0,0,0]
for index, row in pathsDF.iterrows():
   pathLengths=(row['ADEJ'],row['BCDEJ'],row['BCFGJ'],row['BCFHJ'], row['BI'])
   CPL=max(pathLengths)
   CPIndex=pathLengths.index(max(pathLengths))
   CP=paths[CPIndex]
   criticalPath.append(CP)
   floatCritcialPathLength= np.append(floatCritcialPathLength,CPL)
   pathCounts[CPIndex]+=1

pathsDF['critialPath']=criticalPath
print(floatCritcialPathLength)
pathsDF['criticalPathLength']=floatCritcialPathLength
print(pathsDF.head(3))
print(paths)
print(pathCounts)
```
